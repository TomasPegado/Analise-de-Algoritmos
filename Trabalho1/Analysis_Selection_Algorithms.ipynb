```python
import matplotlib.pyplot as plt

# Data from the results
n = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
linear_times = [0.000376, 0.000695, 0.000981, 0.001273, 0.001609, 0.001998, 0.002408, 0.003460, 0.003698, 0.003558]
sort_times = [0.037461, 0.145178, 0.330858, 0.585487, 0.926231, 1.375125, 1.876071, 2.461091, 3.118761, 3.879214]

plt.figure(figsize=(10, 6))
plt.plot(n, linear_times, marker='o', label='LinearSelection (Median of Medians)')
plt.plot(n, sort_times, marker='s', label='SortSelection (BubbleSort)')
plt.xlabel('Input Size (n)')
plt.ylabel('Average Execution Time (s)')
plt.title('Execution Time vs Input Size for Selection Algorithms')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
"""
Comparison with Theoretical Complexity:
- LinearSelection (Median of Medians) is expected to run in O(n) time. The plot shows a nearly flat, very slow-growing line, consistent with linear complexity.
- SortSelection (BubbleSort) is O(n^2), and the plot shows a steep, quadratic-like increase in execution time as n grows, matching the theoretical expectation.
Thus, the empirical results align well with the theoretical time complexities of the algorithms.
""" 
```
